```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import random, time

# 1. Chrome driver yo'lini belgilang (masalan):
driver = webdriver.Chrome()

# 2. eMaktab jurnal sahifasiga kiring (avval tizimga kirgan bo'lish kerak)
driver.get("https://emaktab.uz/journal/school/1000018403503/group/240403630414669932/subject/2265943605112607788")

time.sleep(10)  # sahifa yuklanishini kutish

# 3. Ustun (masalan, 11.09.2025) ichidagi kataklarni topamiz
cells = driver.find_elements(By.CSS_SELECTOR, "td[role='gridcell'] input")

# 4. Faqat bo‘sh kataklarni ajratamiz
empty_cells = [cell for cell in cells if cell.get_attribute("value") == ""]

# 5. 50% ni tanlaymiz va baholarni 4-5 orasida tasodifiy belgilaymiz
fill_count = len(empty_cells) // 2
selected_cells = random.sample(empty_cells, fill_count)

for cell in selected_cells:
    mark = random.choice(["4", "5"])
    cell.send_keys(mark)
    time.sleep(0.3)  # har bir kiritish orasida ozgina kutish

print("✅ 50% kataklar 4 va 5 bilan to‘ldirildi!")
```
