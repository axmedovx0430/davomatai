```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# 1. MNIST ma'lumotlarini yuklaymiz
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255

# 2. CNN modelini yaratamiz
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),#32 ta filtr yordamida 3x3 o‘lchamli xususiyatlarni topadi (chekka, egri chiziq, shakl).
    MaxPooling2D(2,2), #
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# 3. Modelni kompilyatsiya va o‘qitish
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))

# 4. Modelni saqlaymiz
model.save("raqam_model.h5")
print("✅ Model tayyor va 'raqam_model.h5' sifatida saqlandi!")
```

```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
from PIL import Image, ImageGrab
import tkinter as tk

# --- CNN modelni yuklash ---
model = load_model("raqam_model.h5")

# --- Asosiy oynani yaratamiz ---
window = tk.Tk()
window.title("🧠 CNN orqali raqamni tanish (real-time)")
window.geometry("400x480")

# --- Canvas (chizish oynasi) ---
canvas = tk.Canvas(window, width=300, height=300, bg='white')
canvas.pack(pady=20)

# --- Natija uchun label ---
pred_label = tk.Label(window, text="Raqam chiz...", font=("Arial", 16))
pred_label.pack(pady=10)

# --- Chizish funksiyasi ---
def draw(event):
    x1, y1 = (event.x - 8), (event.y - 8)
    x2, y2 = (event.x + 8), (event.y + 8)
    canvas.create_oval(x1, y1, x2, y2, fill='black', width=0)

canvas.bind("<B1-Motion>", draw)

# --- Canvasni tozalash ---
def clear_canvas():
    canvas.delete("all")
    pred_label.config(text="Raqam chiz...")

# --- Model orqali tanish ---
def recognize_digit():
    # Canvasdagi rasmni olish
    x = window.winfo_rootx() + canvas.winfo_x()
    y = window.winfo_rooty() + canvas.winfo_y()
    x1 = x + canvas.winfo_width()
    y1 = y + canvas.winfo_height()

    img = ImageGrab.grab().crop((x, y, x1, y1)).convert('L')
    img = img.resize((28, 28))
    img_array = np.array(img)
    img_array = 255 - img_array  # invert ranglar
    img_array = img_array / 255.0
    img_array = img_array.reshape(1, 28, 28, 1)

    # Model orqali taxmin
    pred = model.predict(img_array)
    digit = np.argmax(pred)
    confidence = np.max(pred) * 100

    # Natijani yangilash
    pred_label.config(text=f"🧠 {digit}  ({confidence:.1f}% ishonch)")

    # 1 soniyadan keyin qayta ishlaydi
    window.after(1000, recognize_digit)

# --- Dastur ishga tushganda avtomatik tanish jarayonini boshlaydi ---
window.after(1000, recognize_digit)

# --- Tugma: Tozalash ---
tk.Button(window, text="Tozalash", command=clear_canvas,
          font=("Arial", 14), bg="#f44336", fg="white").pack(pady=10)

window.mainloop()
```
